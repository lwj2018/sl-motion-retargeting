```python
import numpy as np
import math
def Rxa(a):
    rot = np.zeros([3,3])
    rot[0,0] = 1
    rot[1,1] = math.cos(a)
    rot[1,2] = -math.sin(a)
    rot[2,1] = math.sin(a)
    rot[2,2] = math.cos(a)
    return rot

def Ryb(b):
    rot = np.zeros([3,3])
    rot[0,0] = math.cos(b)
    rot[0,2] = math.sin(b)
    rot[1,1] = 1
    rot[2,0] = -math.sin(b)
    rot[2,2] = math.cos(b)
    return rot

def Rzc(c):
    rot = np.zeros([3,3])
    rot[0,0] = math.cos(c)
    rot[0,1] = -math.sin(c)
    rot[1,0] = math.sin(c)
    rot[1,1] = math.cos(c)
    rot[2,2] = 1
    return rot

def Txyz(xyz):
    x,y,z = xyz
    T = np.identity(4)
    T[0,3] = x
    T[1,3] = y
    T[2,3] = z
    return T

def rpy2rot(rpy):
    c, b, a = rpy
    rot = np.zeros([3,3])
    rot[0][0] = math.cos(a)*math.cos(b)
    rot[0][1] = math.cos(a)*math.sin(b)*math.sin(c)-math.sin(a)*math.cos(c)
    rot[0][2] = math.cos(a)*math.sin(b)*math.cos(c)+math.sin(a)*math.sin(c)
    rot[1][0] = math.sin(a)*math.cos(b)
    rot[1][1] = math.sin(a)*math.sin(b)*math.sin(c)+math.cos(a)*math.cos(c)
    rot[1][2] = math.sin(a)*math.sin(b)*math.cos(c)-math.cos(a)*math.sin(c)
    rot[2][0] = -math.sin(b)
    rot[2][1] = math.cos(b)*math.sin(c)
    rot[2][2] = math.cos(b)*math.cos(c)
    return rot

def expand_as_4x4(rot):
    T = np.identity(4)
    T[:3,:3] = rot
    return T

theta1 = 0.6; theta2 = 1.2;theta3 = 0.1;
theta4 = 1.5; theta5 = 0.2;theta6 = 1.2; theta7 = 2.3;
xyz = [0.05355, -0.0725, 0.51492]
T1 = Txyz(xyz)
rpy = [-0.9795,  -0.5682,   -2.3155]
rot = rpy2rot(rpy)
R1 = np.matmul(rot,Rzc(theta1))
R1 = expand_as_4x4(R1)
T1 = np.matmul(T1,R1)

xyz = [0.03, 0.0, 0.1]
T12 = Txyz(xyz)
R12 = np.matmul(Rxa(1.57079632679),Rzc(theta2))
R12 = expand_as_4x4(R12)
T12 = np.matmul(T12,R12)
T2 = np.matmul(T1,T12)

xyz = [-0.03, 0.17283, 0.0]
T23 = Txyz(xyz)
R23 = np.matmul(Rxa(-1.57079632679),Rzc(theta3))
R23 = expand_as_4x4(R23)
T23 = np.matmul(T23,R23)
T3 = np.matmul(T2,T23)

xyz = [-0.04188, 0.0, 0.07873]
T34 = Txyz(xyz)
R34 = np.matmul(Ryb(-1.57079632679),Rxa(1.57079632679))
R34 = np.matmul(R34,Rzc(theta4))
R34 = expand_as_4x4(R34)
T34 = np.matmul(T34,R34)
T4 = np.matmul(T3,T34)
T4

xyz = [0.0405, 0.16461, 0.0]
T45 = Txyz(xyz)
R45 = np.matmul(Rxa(-1.57079632679),Rzc(theta5))
R45 = expand_as_4x4(R45)
T45 = np.matmul(T45,R45)
T5 = np.matmul(T4,T45)

xyz = [-0.027, 0, 0.10039]
T56 = Txyz(xyz)
R56 = np.matmul(Rxa(1.57079632679),Rzc(theta6))
R56 = expand_as_4x4(R56)
T56 = np.matmul(T56,R56)
T6 = np.matmul(T5,T56)

xyz = [0.027, 0.029, 0.0]
T67 = Txyz(xyz)
R67 = np.matmul(Rxa(-1.57079632679),Rzc(theta7))
R67 = expand_as_4x4(R67)
T67 = np.matmul(T67,R67)
T7 = np.matmul(T6,T67)
T4
```

```python
# pi_2= 1.57079632679
# xyz = [0.05355, -0.0725, 0.51492]
# rpy = [-0.9795,  -0.5682,   -2.3155]
# self.T1 = Txyz(xyz)
# self.R1_static = rpy2rot(rpy)
# self.T1_static = torch.matmul(self.T1,self.R1_static)

# xyz = [0.03, 0.0, 0.1]
# self.T12 = Txyz(xyz)
# self.R12_static = Rxa(pi_2)
# self.T12_static = torch.matmul(self.T12,self.R12_static)

# xyz = [-0.03, 0.17283, 0.0]
# self.T23 = Txyz(xyz)
# self.R23_static = Rxa(-pi_2)
# self.T23_static = torch.matmul(self.T23,self.R23_static)

# xyz = [-0.04188, 0.0, 0.07873]
# self.T34 = Txyz(xyz)
# self.R34_static = Rxa(-pi_2)
# self.T34_static = 
import torch

class TransformMatrix:
    def __init__(self,inputData):
        """
            init with a 4 x 4 list
        """
        self.matrixData = inputData

    @classmethod
    def init_with_tensor(self,tensor:torch.Tensor):
        """
            init with a tensor with the shape of (... x 4 x 4)
        """
        assert tensor.size()[-1] == 4
        assert tensor.size()[-2] == 4
        inputData = [ [None for _ in range(4)] for _ in range(4)]
        for i in range(4):
            for j in range(4):
                inputData[i][j] = tensor[...,i,j]
        print(inputData)
        return TransformMatrix(inputData)

    @classmethod
    def init_with_identity(self):
        """
            init with an Identity tensor
        """
        identity = torch.eye(4,4)
        return TransformMatrix.init_with_tensor(identity)

    @classmethod
    def init_with_axis_angle(self,angle,axis='z'):
        """
            init with axis and angle
            angle(radius) is a Tensor with shape of N x T
        """
        inputData = [ [None for _ in range(4)] for _ in range(4)]
        for i in range(4):
            for j in range(4):
                inputData[i][j] = torch.zeros(angle.size()) 
        if axis == 'x':
            inputData[0][0] = torch.ones(angle.size())
            inputData[1][1] = torch.cos(angle)
            inputData[1][2] = -torch.sin(angle)
            inputData[2][1] = torch.sin(angle)
            inputData[2][2] = torch.cos(angle)
        elif axis == 'y':
            inputData[0,0] = torch.cos(angle)
            inputData[0,2] = torch.sin(angle)
            inputData[1,1] = torch.ones(angle.size())
            inputData[2,0] = -torch.sin(angle)
            inputData[2,2] = torch.cos(angle)
        elif axis == 'z':
            inputData[0][0] = torch.cos(angle)
            inputData[0][1] = -torch.sin(angle)
            inputData[1][0] = torch.sin(angle)
            inputData[1][1] = torch.cos(angle)
            inputData[2][2] = torch.ones(angle.size())
        return TransformMatrix(inputData)

    def __mul__(self,other): 
        if isinstance(other,TransformMatrix):
            result = [ [None for _ in range(4)] for _ in range(4)]
            for i in range(4):
                for j in range(4):
                    result[i,j] = TransformMatrix.mul_row_col(self.matrixData,other.matrixData,i,j)
            return TransformMatrix(result)
        if isinstance(other,torch.Tensor):
            result = [ [None for _ in range(4)] for _ in range(4)]
            for i in range(4):
                for j in range(4):
                    result[i][j] = TransformMatrix.mul_row_col_with_tensor(self.matrixData,other,i,j)
            return TransformMatrix(result)
        raise TypeError('Cannot multiply/add transfromMatices with type %s' % str(type(other)))

    @classmethod
    def mul_row_col(cls,x,y,i,j):
        """
            multiply x's i-th row with y's j-th column
            both x & y are TransformMatrices
            TODO: support broadcast
        """
        result = x[i][0]*y[0][j] + x[i][1]*y[1][j]  + x[i][2]*y[2][j] + x[i][3]*y[3][j] 
        return result

    @classmethod
    def mul_row_col_with_tensor(cls,x,y,i,j):
        """
            multiply x's i-th row with y's j-th column
            y is a Tensor with shape of (... x 4 x 4)
        """
        result = x[i][0]*y[...,0,j] + x[i][1]*y[...,1,j]  + x[i][2]*y[...,2,j] + x[i][3]*y[...,3,j] 
        return result

    @property
    def position(self):
        data = self.matrixData
        return torch.stack([data[0][3],data[1][3],data[2][3]],dim=-1)

    @property
    def transform(self):
        data = self.matrixData
        row0 = torch.stack([data[0][0],data[0][1],data[0][2],data[0][3]],dim=-1)
        row1 = torch.stack([data[1][0],data[1][1],data[1][2],data[1][3]],dim=-1)
        row2 = torch.stack([data[2][0],data[2][1],data[2][2],data[2][3]],dim=-1)
        row3 = torch.stack([data[3][0],data[3][1],data[3][2],data[3][3]],dim=-1)
        return torch.stack([row0,row1,row2,row3],dim=-2)
```
